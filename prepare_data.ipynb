```python
dr_ = df[df['damage'] == 'DR']
g_ = df[df['damage'] == 'G']
nd_ = df[df['damage'] == 'ND']
wd_ = df[df['damage'] == 'WD']
other_ = df[df['damage'] == 'other']
```

```python
from sklearn.model_selection import train_test_split
x_test, x_train = train_test_split(dr_, test_size=0.2)
```

```python
for i in damage.keys():
    os.mkdir(f'./datasets/val/{damage[i]}')
```

```python
src = './datasets/images/'
des = './datasets/'

for df_ in [dr_, g_, nd_, wd_, other_]:
    train, val = train_test_split(df_, test_size=0.2)
    print(df_['damage'].tolist()[0])
    [shutil.copy(os.path.join(src, file), os.path.join(des, 'train', df_['damage'].tolist()[0], file)) for file in train['filename']]
    [shutil.copy(os.path.join(src, file), os.path.join(des, 'val', df_['damage'].tolist()[0], file)) for file in val['filename']]

```
